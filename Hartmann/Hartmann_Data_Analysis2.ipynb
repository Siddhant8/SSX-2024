Run this with Ha = 1 and move the scratch directory to scratch_Bz_1_Ha_1, then run again with Ha=10 to recreate the following plot

```python
df_Ha1 = h5py.File("scratch_Bz_1_Ha_1/integrals/integrals_s1.h5","r")
df_Ha20 = h5py.File("scratch/integrals/integrals_s1.h5","r")
```

```python
z = df_Ha20['tasks/<vx>_xy'].dims[3][0][:]
```

```python
vx_Ha20 = df_Ha20['tasks/<vx>_xy'][-1,0,0,:]
vx_Ha1 = df_Ha1['tasks/<vx>_xy'][-1,0,0,:]
plt.plot(z, vx_Ha1/vx_Ha1.max(), label=r'$\mathrm{Ha} = 1$')
plt.plot(z, vx_Ha20/vx_Ha20.max(), label=r'$\mathrm{Ha} = 20$')
plt.plot(z, 1- 4*(0.5-z)**2, label=r'$B_z=0$ (hydro)')
plt.legend()
plt.xlabel("z")
plt.ylabel(r"$\left< v_x \right>/v_x(z=1)$")
```
